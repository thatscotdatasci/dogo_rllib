```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rc('font', size=16)
```

```python
RESULTS_BASEDIR = "../dogo_results/mopo"
get_results_path = lambda x: os.path.join(RESULTS_BASEDIR, x) if os.path.isfile(os.path.join(RESULTS_BASEDIR, x)) else x
```

# Initial Run

```python
INITIAL = get_results_path("ray-rllib/halfcheetah-ppo/PPO_HalfCheetah-v2_ac579_00000_0_2022-05-26_14-35-16/result.json")
initial_df = pd.read_json(INITIAL, lines=True)
```

```python
checkpoints_index = np.arange(10, 1390, 10)-1
feature = 'episode_reward_mean'
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.plot(initial_df['timesteps_total'], initial_df[feature], label='RLLIB_PPO_001')
ax.plot(initial_df.iloc[checkpoints_index]['timesteps_total'], initial_df.iloc[checkpoints_index][feature], label='RLLIB_PPO_001 - Checkpoints')
ax.set_xlabel('Steps')
ax.set_ylabel(feature)
ax.legend()
pass
```

```python
initial_df.iloc[checkpoints_index][['timesteps_total', 'episode_reward_mean']]
```
